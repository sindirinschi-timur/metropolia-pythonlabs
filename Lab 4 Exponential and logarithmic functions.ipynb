## Problem 2.a

The voltage of a capacitor decays exponentially (See Software 1, Mathematics Lectures Period 1.pdf, page 27):

$ y(t) = \begin{cases} V_0,\; t<0 \\ V_0 \cdot e^{-t/RC}, \; t \geq 0 \end{cases}$

Sketch the graph of the voltage of a capacitor in time-range -2 ms to 5 ms (included) using timestep of $0.01$ ms (Note: 1.0 ms = 0.001 s). Use the following values in calculations:
- $V_0 = 5.0$
- $R = 1.2 \times 10^3$
- $C = 1.0 \times 10^{-6}$

Draw in the same graph the following function

$ y_2(t) = 1.5 \cdot \left( y(t) > 1.5 \right)$

Annotate the graph (xlabel, ylabel, title, and grid-lines). 

````python
V0 = 5.0
R = 1.2e3
C = 1.0e-6

t = np.arange(-2e-3, 5e-3, 0.01e-3)
y = np.where(t < 0, V0, V0 * np.exp(-t / (R * C)))
y2 = 1.5 * (y > 1.5)

plt.figure(figsize=(8, 6))
plt.plot(t * 0.001e3, y, label='Voltage decay y(t)', color='blue')
plt.plot(t * 0.001e3, y2, label='$y_2(t)$', color='red', linestyle='--')
plt.axhline(0, linewidth=0.5)
plt.xlabel('Time (ms)')
plt.ylabel('Voltage (V)')
plt.title('Voltage Decay in a Capacitor')
plt.grid(True)
plt.legend()
plt.show()
````

## Problem 3.a

The temperature of a hot cup of coffee decreases according to the formula (See Software 1, Mathematics Lectures, page 27):

$ T(t) = T_{env} + (T_0 - T_{env})\cdot e^{-kt} $

where,
- $T(t)$ is the temperature of the cup of coffee
- $T_{env}$ is the temperature of the environment
- $T_0$ is the initial temperature of the cup of coffee
- $k$ is the time constant
- $t$ is time

Sketch a graph of $T(t)$, when $T_{env} = 25.0$, $T_0 = 90.0$, and $k = 0.02$. Use time range from 0.0 s to 180.0 s having 200 *linearly spaced* values.

Annotate the graph (grid-lines, xlabel, ylabel, title).

````python
T_env = 25.0
T_0 = 90.0
k = 0.02
t = np.linspace(0, 180, 200)

T = T_env + (T_0 - T_env) * np.exp(-k * t)

plt.figure(figsize=(8, 6))
plt.plot(t, T, label=r'Temperature $T(t)$', color='red')
plt.axhline(45, color='blue', linestyle='--', label='Under 45C (Undrinkable)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (C)')
plt.title('Coffee Cooling')
plt.grid(True)
plt.legend()
plt.show()
````

## Problem 4

Sketch the graphs of 
- $f_1(x) = x$
- $f_2(x) = e^{x}$
- $f_3(x) = ln(x)$

using three separate axis.

Draw 
- $f_1(x)$ in the first axis using lin-lin scale (`plt.plot()`)
- $f_2(x)$ in the second axis using lin-log scale (`plt.semilogy()`)
- $f_3(x)$ in the third axis using log-lin scale(`plt.semilogx()`).

Use *logarithmically spaced values* for x to draw the graphs (e.g., `x = np.logspace(-9, 0.7, 500)`)

Add grid-lines to all graphs, Use labels and legends to annotate the graphs.

What do you observe? Explain it.

````python
x = np.logspace(-9, 0.7, 500)
f1 = x
f2 = np.exp(x)
f3 = np.log(x)

fig, ax = plt.subplots(3, figsize=(6, 12))
ax[0].plot(x, f1, label=r'$f_1(x) = x$', color='black')
ax[0].set_xscale('linear')
ax[0].set_yscale('linear')
ax[0].grid(True)
ax[0].set_title('lin-lin scale')
ax[0].legend()
ax[1].semilogy(x, f2, label=r'$f_2(x) = e^x$', color='blue')
ax[1].grid(True)
ax[1].set_title('lin-log scale')
ax[1].legend()
ax[2].semilogx(x, f3, label=r'$f_3(x) = ln(x)$', color='red')
ax[2].grid(True)
ax[2].set_title('log-lin scale')
ax[2].legend()

plt.tight_layout()
plt.show()

print('Observations: \nExponential functions grow rapidly if x is large, while logarithmic functions will grow slowly.')
print('It is important to choose the right scale in concordance with our particular data so we can observe and analyze trends better.')
````

## Problem 5

In 1965, **[Gordon Moore](https://en.wikipedia.org/wiki/Gordon_Moore)** was asked by Electronics Magazine to predict what might happen in the semiconductor components over the next ten years. Moore observed that the numbers of components (transistors, etc.) in a dense integrated circuit had doubled approximately every year and speculated it would continue to do so for at least the next ten years. This forecast rate has been revised since then but the basic principle has continued over 50 years. 

Now it is your turn to make predictions. The cleaned data in file `transistors.dat` is downloaded from **[What is Moore's Law? - Our World in Data](https://ourworldindata.org/moores-law)**. You can read the data in to Notebook with the following code:
```python
data = np.genfromtxt('transistors.dat', delimiter = '  ')
year = data[:, 0]
transistors = data[:, 1]
```

Create two separate figures. Use `plt.figure()` commands. 

Draw a scatter plot of the data in the first figure using lin-lin scale, e.g. `plt.scatter(year, transistors)`. Decorate the graph, e.g. add xlabel, ylabel, title, grid-lines and so on. It is difficult to predict the output from this graph, but it serves as a linear reference. Leave it as it is and continue with the second figure.

Draw a second scatter plot of the data in the second figure, but now plot the ten-base logarithmic values for the transistors, e.g. `plt.scatter(year, np.log10(transistors)`. Decorate also this graph.

Try to find the linear relationship between the ten-base logarithmic values of the transistors and the years, e.g. $y(x) = k\cdot x + b$, where $x$ is the year, $k$ is the slope, $b$ is the offset, and $y$ is the estimate of `np.log10(transistors)`. Draw this estimate line in the same graph with the data, e.g. use `plt.plot(x, y)`. Use *linearly spaced values* for $x \in [1970, 2030]$. Change the xlimits and ylimits to make the graphics more readable.

Find the value of $y(2030)$. Convert that back to linear scale by calculating `N = 10**y`. This is your prediction of the number of transistors in ICs at year 2030. 

What is your prediction?

````python
import numpy as np
import matplotlib.pyplot as plt

# Read in the data
data = np.genfromtxt('transistors.dat', delimiter='  ')
year = data[:, 0]
transistors = data[:, 1]

# Draw the lin-lin scatter plot
fig1 = plt.figure(figsize=(8, 6))
plt.scatter(year, transistors, color='blue', label='Transistors')
plt.xlabel('Year')
plt.ylabel('Transistor Amount')
plt.title('Transistor Count (lin-lin scale)')
plt.grid(True)
plt.legend()
plt.show()

# Draw the lin-log10 scatter plot
fig2 = plt.figure(figsize=(8, 6))
log_transistors = np.log10(transistors)
plt.scatter(year, log_transistors, color='red', label='Log10 Transistors')
plt.xlabel('Year')
plt.ylabel('Log10(Transistor Amount)')
plt.title('Logarithmic Transistor Count')
plt.grid(True)
plt.legend()
c = np.polyfit(year, log_transistors, 1)
k, b = c
print(f"Slope (k) is {k}")
print(f"Intercept (b) is {b}")
x_fit = np.linspace(1970, 2030, 100)
y_fit = k * x_fit + b
plt.plot(x_fit, y_fit, color='black', linestyle='--', label=f'y={k:.4f}x + {b:.2f}')
plt.xlim(1970, 2030)
plt.ylim(min(log_transistors)-1, max(log_transistors)+1)
plt.legend()
plt.show()
y_2030 = k * 2030 + b
N = y_2030 ** 10
print(f"Prediction for 2030 is {N:.2e} transistors")
````
